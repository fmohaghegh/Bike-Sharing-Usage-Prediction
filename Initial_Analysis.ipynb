# Prediction of Bike Sharing System Usage

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading the data:

```python
alldata = pd.read_csv('hour.csv')
alldata.head()
```

Every day varaion:

```python
alldata.iloc[:].plot(x='instant',y='cnt')
plt.xlabel('instant')
plt.ylabel('count')
```

```python
alldata.iloc[:24*7].plot(x='dteday',y='cnt')
plt.xlabel('date')
plt.ylabel('count')
```

Updating the X as the input:

```python
df  =  alldata.drop(['dteday','casual','registered'],axis=1)

[n_days,n_params]  = np.shape(df)

print('[n_days,n_params] = ',[n_days,n_params])
```
